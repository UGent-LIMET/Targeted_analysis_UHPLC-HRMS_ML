```python
# Importing Image and ImageOps module from PIL package
from PIL import Image, ImageOps
import os

wd = ('/media/sf_SF/Stage2021/targetedQE/data/input/MachineLearning/Xarrays/') #png's
output = ('/media/sf_SF/Stage2021/targetedQE/data/input/MachineLearning/Xarrays_bw/') #output

os.chdir(wd)
list_png = os.listdir(wd)
print(list_png)

for png in list_png :
    # creating an og_image object
    og_image = Image.open(png)
    # applying grayscale method
    gray_image = ImageOps.grayscale(og_image)
    new = png.split(".png")[0]
    gray_image.save(output + new + '.png')
```
