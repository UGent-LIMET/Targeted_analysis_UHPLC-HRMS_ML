```python

############Adjustments##############

#options
PATH = '/media/sf_SF/Stage2021/targetedQE/' 
OPTION = 'label' # area or label
FOLDER = 'X_arrays_7377/'

## Adjustments

filename_Y_labels = 'total_y_matrix.txt'

########################



# load libraries
import pandas as pd
import os
import random
from PIL import Image
import numpy as np


#set paths
path_data_in = PATH + 'data/input/' + 'MachineLearning/'
path_data_out = PATH + 'data/output/' + 'MachineLearning/'
path_data_X = path_data_in + FOLDER #png's
path_data_y = path_data_in + 'Yarrays/' #labels


filename = path_data_y + filename_Y_labels

y = pd.read_csv(filename, sep = "\t")

filenames_divide = []
directory_list = os.listdir(path_data_X)
random.shuffle(directory_list)
os.chdir(path_data_X)

for filename in directory_list:
    if ".png" in filename :
        filenames_divide.append(path_data_X + filename)
    
def divide_f_nf(filenames_list, y):

    found_images_as_array=[]
    not_found_images_as_array=[]
    ordered_filenames_found = pd.DataFrame()
    ordered_filenames_nf = pd.DataFrame()
    
    
    for filename in filenames_list:
        filename_wopath = filename.split(FOLDER)[1]
        filename_wopath = filename_wopath.strip(".png")

        matching_y = y[y.SampleTargetedcombination==filename_wopath]
        if len(matching_y) == 1:
            
            ar = matching_y.iloc[0,3]
            lab = matching_y.iloc[0,5]
            
            if lab == 'FOUND':
                found_lab_area = pd.DataFrame(np.array([[filename_wopath,lab,ar]]))
                ordered_filenames_found = ordered_filenames_found.append(found_lab_area, ignore_index=True)
                
            else:
                nf_lab_area = pd.DataFrame(np.array([[filename_wopath,lab,ar]]))
                ordered_filenames_nf = ordered_filenames_nf.append(nf_lab_area, ignore_index=True)
            
            
        if len(matching_y) != 1:
            continue
    
         
    return np.array(ordered_filenames_nf), np.array(ordered_filenames_found)
    
not_found, found = divide_f_nf(filenames_divide,y)


def divide_f_nf(filenames_list):
    list_train = []
    list_test = []
    i = 0
    for filename in filenames_list:
        if i % 3 == 0: 
            list_test.append(filename)
        else:
            list_train.append(filename)
        i = i + 1
    return list_train, list_test

filenames_X_train_f, filenames_X_test_f = divide_f_nf(found)
filenames_X_train_nf, filenames_X_test_nf = divide_f_nf(not_found)


def load_y(filenames_list, y):
    all_images_as_array = []
    area = []
    global ordered_filenames
    ordered_filenames=pd.DataFrame()
    
    for filename in filenames_list :
        filename_wopath = filename[0]

        img=Image.open(path_data_X + filename_wopath + ".png")
        np_array = np.asarray(img)

        l,b,c = np_array.shape    
        np_array = np_array.reshape(l*b*c,)   
        all_images_as_array.append(np_array)

        area.append(filename[2])
        
        file_area = pd.DataFrame(np.array([[filename_wopath,filename[2]]]))
        ordered_filenames = ordered_filenames.append(file_area, ignore_index=True)
            

    return np.array(all_images_as_array), area, ordered_filenames


X_test_nf, y_test_nf, test_nf = load_y(filenames_X_test_nf,y)
X_test_f, y_test_f, test_f = load_y(filenames_X_test_f,y)
X_train_nf, y_train_nf, train_nf = load_y(filenames_X_train_nf,y)
X_train_f, y_train_f, train_f = load_y(filenames_X_train_f,y)



print(len(X_test_nf))
print(y_test_nf)

print(len(X_test_f))
print(y_test_f)

print(len(X_train_nf))
print(y_train_nf)

print(len(X_train_f))
print(y_train_f)
```
